# Вычисления с помощью Numpy¶

```python
# Задание 1
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 


x = [i  for i in range(1, 8)]
y = [ 3.5, 3.8, 4.2, 4.5, 5, 5.5, 7, ]

plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# Задание 2
t = np.linspace(0, 10, 51)
f = np.array([np.cos(i) for i in t])

plt.plot(t, f, color = 'g')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.title('График f(t)', fontsize=20)
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show()
```

```python
# Задание 3
x = np.linspace(-3, 3, 51)
y1 = np.array([i**2 for i in x])
y2 = np.array([2 * i + 5 for i in x])
y3 = np.array([-3 * i - 1.5 for i in x])
y4 = np.array([np.sin(i) for i in x])

Y = [y1, y2, y3, y4]

fig = plt.figure(figsize = (8,6))

ax = [fig.add_subplot(2, 2, i) for i in range(1,5)]
print(ax)

for i in range(len(ax)):
    ax[i].set(title = f'График у{i + 1}')
    ax[i].plot(x, Y[i])
    if i == 0:
        ax[i].set_xlim([-5, 5])
        
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
